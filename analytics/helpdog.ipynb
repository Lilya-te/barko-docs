```python
import pandas as pd
```

```python
df = pd.read_csv('forum_data.csv')
```

```python
df.head()
```

```python
print('Количество собак на форуме:', df['link'].unique().shape[0])
```

```python
df_home = df[~df['home_started_at'].isnull()].copy()
print('Количество собак, которые были пристроены 1 или более раз:', df_home['link'].unique().shape[0])
```

```python
# удаляем записи с пустым полем "дата возвращения в приют", оставляем только последнее пристройство без даты "дата возвращения в приют"
for link in df_home['link'].value_counts().items():
    dog_hist = df[df['link'] == link[0]].copy()
    last_home_date = dog_hist['home_started_at'].max()
    inx_to_drop = dog_hist[(dog_hist['back_shelter_at'].isnull()) & (dog_hist['home_started_at'] != last_home_date)].index.values
    df_home.drop(index=inx_to_drop, inplace=True)
```

```python
df_home
```

```python
print('Вернули в приют 1 или более раз: ', df_home[~df_home['back_shelter_at'].isnull()]['link'].unique().shape[0])
print('Процент вернувшихся в приют после пристройства:', round(df_home[~df_home['back_shelter_at'].isnull()]['link'].unique().shape[0]/df_home['link'].unique().shape[0]*100), '%')
```
