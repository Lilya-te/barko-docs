```python
# -----------------------------
# Diagram 3: Wireframe Prototype
# -----------------------------
from html import escape


def card(title, *lines):
    items = "\n".join(
        f'          <TR><TD ALIGN="LEFT" BALIGN="LEFT">&#8226;&#160;{escape(line)}</TD></TR>'
        for line in lines
    )
    return f"""<
        <TABLE BORDER="0" CELLBORDER="0" CELLSPACING="0" CELLPADDING="2">
          <TR><TD ALIGN="CENTER"><B>{escape(title)}</B></TD></TR>
          <TR><TD ALIGN="CENTER"><FONT COLOR="#475569">- - - - - - - - - -</FONT></TD></TR>
{items}
        </TABLE>
        >"""

with g.subgraph(name="cluster_wireframe") as c:
    c.attr(
        fontsize="24",
        fontname="Helvetica-Bold",
        fontcolor="#334155",
        color="transparent",
        penwidth="0",
        style="rounded",
        fillcolor="transparent",
        margin="22",
    )

    c.node(
        "w1",
        card("Вход", "Magic link", "Согласие на обработку данных"),
        fillcolor="#E0F2FE",
        color="#38BDF8",
    )
    c.node(
        "w2",
        card("Питомец", "Возраст, порода", "Особенности", "Фото"),
        fillcolor="#DBEAFE",
        color="#60A5FA",
    )
    c.node(
        "w3",
        card("Оценка (ML модель)", "Risk score", "Сложные зоны", "Объяснение весов"),
        fillcolor="#DCFCE7",
        color="#4ADE80",
    )
    c.node(
        "w6",
        card("Оплата", "Подписка", "7 / 30 / 90 дней"),
        fillcolor="#CCFBF1",
        color="#2DD4BF",
    )
    c.node(
        "w4",
        card("План", "День 1-7 / 30 / 90", "Задачи", "Напоминания"),
        fillcolor="#FEF3C7",
        color="#F59E0B",
    )
    c.node(
        "w5",
        card("Чат-бот (LLM + RAG)", "Вопрос", "Ответ по контексту"),
        fillcolor="#FCE7F3",
        color="#F472B6",
    )
    c.node(
        "w7",
        card("База знаний", "Статьи", "Практические советы"),
        fillcolor="#E0F2FE",
        color="#38BDF8",
    )

    # Main product flow: left to right.
    c.edge("w1", "w2")
    c.edge("w2", "w3")
    c.edge("w3", "w6")
    c.edge("w1", "w7")

    # Paid features branch into a compact right-hand column.
    with c.subgraph(name="cluster_features") as features:
        features.attr(rank="same", style="invis")
        features.node("w4")
        features.node("w5")

    c.edge("w6", "w4", xlabel="доступ открыт")
    c.edge("w6", "w5", xlabel="доступ открыт")
    c.edge("w4", "w5", style="invis", weight="20", constraint="false")

# Global title
g.attr(
    labelloc="t",
    fontsize="30",
    fontname="Helvetica-Bold",
    fontcolor="#0F172A",
)

g.render("pet_adaptation_architecture_slide", cleanup=True)
print("Saved: pet_adaptation_architecture_slide.svg")
```

```python
import xml.etree.ElementTree as ET
from pathlib import Path

def html(text: str) -> str:
    return text.replace("\n", "<br>")

def add_cell(root, id_, value, x, y, w, h, fill="#F8FAFC", stroke="#CBD5E1"):
    cell = ET.SubElement(root, "mxCell", {
        "id": id_,
        "value": html(value),
        "style": (
            "rounded=1;whiteSpace=wrap;html=1;"
            f"fillColor={fill};strokeColor={stroke};"
            "fontFamily=Arial;fontSize=14;align=center;verticalAlign=middle;"
        ),
        "vertex": "1",
        "parent": "1",
    })
    ET.SubElement(cell, "mxGeometry", {
        "x": str(x),
        "y": str(y),
        "width": str(w),
        "height": str(h),
        "as": "geometry",
    })

def add_edge(root, id_, source, target, label=""):
    edge = ET.SubElement(root, "mxCell", {
        "id": id_,
        "value": label,
        "style": (
            "endArrow=block;html=1;rounded=0;"
            "strokeColor=#64748B;fontFamily=Arial;fontSize=12;"
        ),
        "edge": "1",
        "parent": "1",
        "source": source,
        "target": target,
    })
    ET.SubElement(edge, "mxGeometry", {
        "relative": "1",
        "as": "geometry",
    })

mxfile = ET.Element("mxfile", {
    "host": "app.diagrams.net",
    "version": "24.7.17",
    "type": "device",
})

diagram = ET.SubElement(mxfile, "diagram", {
    "id": "pet-architecture",
    "name": "Architecture",
})

model = ET.SubElement(diagram, "mxGraphModel", {
    "dx": "1600",
    "dy": "900",
    "grid": "1",
    "gridSize": "10",
    "guides": "1",
    "tooltips": "1",
    "connect": "1",
    "arrows": "1",
    "fold": "1",
    "page": "1",
    "pageScale": "1",
    "pageWidth": "1600",
    "pageHeight": "900",
    "math": "0",
    "shadow": "0",
})

root = ET.SubElement(model, "root")
ET.SubElement(root, "mxCell", {"id": "0"})
ET.SubElement(root, "mxCell", {"id": "1", "parent": "0"})

add_cell(root, "title", "Архитектура и прототип сервиса адаптации питомца", 360, 20, 880, 50, "#FFFFFF", "#FFFFFF")

# 1. Архитектура технического решения
add_cell(root, "d1", "1. Архитектура технического решения", 40, 100, 460, 40, "#E0F2FE")
add_cell(root, "app", "Mobile App\nReact Native / Flutter", 80, 170, 180, 70, "#DBEAFE")
add_cell(root, "auth", "K1 Auth + PII\nSupabase / Firebase Auth", 300, 150, 180, 70, "#E0F2FE")
add_cell(root, "api", "Backend API\nFastAPI / Node.js", 300, 260, 180, 70, "#DCFCE7")
add_cell(root, "storage", "Object Storage\nS3-compatible", 540, 130, 180, 70, "#FEF3C7")
add_cell(root, "db", "PostgreSQL\nprofiles, plans, payments", 540, 230, 180, 70, "#FDE68A")
add_cell(root, "vector", "pgvector / Vector DB\nexpert scenarios", 540, 330, 180, 70, "#FCE7F3")
add_cell(root, "llm", "Commercial LLM API\nOpenAI / Anthropic", 780, 330, 180, 70, "#EDE9FE")
add_cell(root, "push", "Notifications\nFCM + APNS / OneSignal", 780, 230, 180, 70, "#FFE4E6")
add_cell(root, "pay", "Payments\nЮKassa / CloudPayments", 780, 130, 180, 70, "#CCFBF1")

add_edge(root, "e1", "app", "auth", "login")
add_edge(root, "e2", "app", "api", "profile / plan / chat")
add_edge(root, "e3", "api", "storage", "files")
add_edge(root, "e4", "api", "db", "data")
add_edge(root, "e5", "api", "vector", "RAG search")
add_edge(root, "e6", "vector", "llm", "context")
add_edge(root, "e7", "api", "push", "reminders")
add_edge(root, "e8", "api", "pay", "access")

# 2. Поток персонализации
add_cell(root, "d2", "2. Поток персонализации продукта", 40, 470, 460, 40, "#E0F2FE")
add_cell(root, "f1", "K1\nРегистрация", 60, 550, 140, 70, "#E0F2FE")
add_cell(root, "f2", "K2\nПрофиль питомца", 230, 550, 140, 70, "#DBEAFE")
add_cell(root, "f3", "K3\nСкоринг рисков", 400, 550, 140, 70, "#DCFCE7")
add_cell(root, "f4", "K4\nПлан адаптации", 570, 550, 140, 70, "#FEF3C7")
add_cell(root, "f5", "K5\nИИ-чат + сценарии", 740, 550, 140, 70, "#FCE7F3")
add_cell(root, "f6", "K6\nМониторинг", 570, 670, 140, 70, "#FFE4E6")
add_cell(root, "f7", "K7\nОплата доступа", 400, 670, 140, 70, "#CCFBF1")

add_edge(root, "e9", "f1", "f2")
add_edge(root, "e10", "f2", "f3")
add_edge(root, "e11", "f3", "f4", "risk score")
add_edge(root, "e12", "f4", "f5", "context")
add_edge(root, "e13", "f4", "f6", "tasks")
add_edge(root, "e14", "f6", "f4", "correction")
add_edge(root, "e15", "f7", "f4", "unlock")

# 3. Проволочный интерфейс
add_cell(root, "d3", "3. Проволочный интерфейс MVP", 1040, 100, 420, 40, "#E0F2FE")
add_cell(root, "w1", "Вход\nMagic link\nСогласие на данные", 1060, 170, 180, 90, "#E0F2FE")
add_cell(root, "w2", "Питомец\nВозраст / порода\nФото / документы", 1280, 170, 180, 90, "#DBEAFE")
add_cell(root, "w3", "Оценка\nRisk score\nОбъяснение весов", 1060, 310, 180, 90, "#DCFCE7")
add_cell(root, "w4", "План\n7 / 30 / 90 дней\nЗадачи и напоминания", 1280, 310, 180, 90, "#FEF3C7")
add_cell(root, "w5", "ИИ-чат\nОтвет по базе\nСовет по плану", 1060, 450, 180, 90, "#FCE7F3")
add_cell(root, "w6", "Оплата\nКарта / СБП\nДоступ открыт", 1280, 450, 180, 90, "#CCFBF1")

add_edge(root, "e16", "w1", "w2")
add_edge(root, "e17", "w2", "w3")
add_edge(root, "e18", "w3", "w4")
add_edge(root, "e19", "w4", "w5")
add_edge(root, "e20", "w6", "w4", "unlock")

tree = ET.ElementTree(mxfile)
ET.indent(tree, space="  ")

output = Path("pet_architecture.drawio")
tree.write(output, encoding="utf-8", xml_declaration=True)

print(f"Saved: {output}")
```
